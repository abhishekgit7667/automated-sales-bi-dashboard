```python
import pandas as pd
from sqlalchemy import create_engine

# 1. Read CSV
df = pd.read_csv("D:/Automated_Sales_BI_Dashboard/data/raw/Superstore.csv", encoding="latin1")

# 2. Clean column names
df.columns = (
    df.columns.str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)

# 3. Convert data types
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
df["ship_date"] = pd.to_datetime(df["ship_date"], errors="coerce")

df["sales"] = pd.to_numeric(df["sales"], errors="coerce")
df["quantity"] = pd.to_numeric(df["quantity"], errors="coerce")
df["discount"] = pd.to_numeric(df["discount"], errors="coerce")
df["profit"] = pd.to_numeric(df["profit"], errors="coerce")

# 4. Create derived columns
df["order_month"] = df["order_date"].dt.month_name()
df["ship_days"] = (df["ship_date"] - df["order_date"]).dt.days
df["profit_margin"] = (df["profit"] / df["sales"]) * 100

df["discount_bucket"] = df["discount"].apply(
    lambda x: "No Discount" if pd.notnull(x) and x == 0 else
              "Low Discount" if pd.notnull(x) and x <= 0.2 else
              "Medium Discount" if pd.notnull(x) and x <= 0.5 else
              "High Discount" if pd.notnull(x) else None
)

```
